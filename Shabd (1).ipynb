```python
import requests
from bs4 import BeautifulSoup
import re

URL = "https://www.shiksha.com/humanities-social-sciences/articles/top-10-colleges-in-india-mhrd-blogId-15781"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup.title
```

```python
t = []

for text in soup.find_all('p'):
    t.append(text.getText())
t
```

```python
h_list = []
heading = ['h1','h2','h3','h4']
for h in soup.find_all(heading):
  h_list.append(h.getText())
h_list
```

```python
table_header = []
for h in soup.find_all('th'):
  table_header.append(h.getText())
table_header

```

```python
a = soup.find_all("table")
table1 = a[66]
body = table1.find_all("tr")
body_rows = body[1:]
all_rows = [] 
for row_num in range(len(body_rows)): 
    row = [] 
    for row_item in body_rows[row_num].find_all("td"): 
        row.append(row_item.getText().strip().split('/n'))
    all_rows.append(row)
all_rows
```

```python
b = soup.find_all("table")
table1 = b[66]
tr_list = []
for tr in table1.find_all('tr'):
  for a in tr.find_all('a', href=True):
    all_text = str(tr.getText().strip()+'\n'+a['href'])
    tr_list.append(all_text.split('\n'))

tr_list
```

```python
import pandas as pd
table_header.append('URL')
df_data = pd.DataFrame(tr_list, columns=table_header)
df_data
```

```python
df_data.to_csv('Shabd_top10collegeinfo.csv',sep='|')
```

```python
b = soup.find_all("table")
table1 = b[66]
body = table1.find_all("tr")
body_rows = body[1:]
url_list = []
for tr in table1.find_all('tr'):
  for a in tr.find_all(href=True):
    all_text =a['href']
    url_list.append(all_text.split('\n'))
a = 0
tv = []
d = []
e = []
def url(lists):
    if len(lists) == 0:
        return lists
    if isinstance(lists[0], list):
        return url(lists[0]) + url(lists[1:])
    return lists[:1] + url(lists[1:])
tv = url(url_list)
arows = [] 
for r_num in range(len(body_rows)): 
    r = [] 
    for r_item in body_rows[r_num].find_all("a"): 
        r.append(r_item.getText().strip().split('/n'))
    arows.append(r)
def arr(clg):
    if len(clg) == 0:
        return clg
    if isinstance(clg[0], list):
        return arr(clg[0]) + arr(clg[1:])
    return clg[:1] + arr(clg[1:])
e = arr(arows)
k = 0
for x in tv:
  r = requests.get(x)
  s = BeautifulSoup(r.content, "html.parser")
  j = e[k]
  d.append(j + ' | ' + x + ' | ')
  for txt in s.find_all("p"):
     d.append(txt.getText())
  '\n'
  k+=1
d
```

```python
import csv
with open('Shabd_CollegeDetails.csv', 'w') as f:
  write = csv.writer(f)
  write.writerow(d)
```
